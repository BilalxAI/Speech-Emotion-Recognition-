Now we define the functions for the waveplot and spectrogram


```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion, size=20)
    librosa.display.waveplot(data, sr=sr)
    plt.show()

def spectogram(data, sr, emotion):
     x = librosa.stft(data)
     xdb = librosa.amplitude_to_db(abs(x))
     plt.figure(figsize=(11,4))
     plt.title(emotion, size=20)
     librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
     plt.colorbar()
```

```python
import librosa
import numpy as np
import zipfile
import os

def extract_mfcc(filename):
    # Check if the file is a ZIP archive
    if zipfile.is_zipfile(filename):
        # Extract the audio file from the ZIP archive
        audio_file = None

        with zipfile.ZipFile(filename, 'r') as zip_ref:
            # Assuming there's only one file in the ZIP archive, adjust as needed
            audio_file = zip_ref.namelist()[0]  # Get the first file in the ZIP

            # Extract to a temporary directory or use Colab's runtime space
            zip_ref.extract(audio_file, path='/content/')

        # Check if audio file was extracted successfully
        if audio_file is None:
            raise ValueError(f"No audio file found in {filename}")

        audio_path = os.path.join('/content/', audio_file)
    else:
        # Directly use the file path as audio_path if it's not a ZIP file
        audio_path = filename

    # Load and process the audio file using librosa
    y, sr = librosa.load(audio_path, duration=3, offset=0.5)
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)

    # Clean up: Remove the extracted audio file if it was from a ZIP
    if zipfile.is_zipfile(filename):
        os.remove(audio_path)

    return mfcc
```

```python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
```

```python
X_mfcc
```

```python
X = [x for x in X_mfcc]
X = np.array(X)
X.shape
```

```python
## input split
X = np.expand_dims(X, -1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
```

```python
y = y.toarray()
```

```python
y.shape
```


**Create the LSTM Model**

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(256, return_sequences=False, input_shape=(40,1)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

**Now we train the model**

```python
# Train the model
history = model.fit(X, y, validation_split=0.2, epochs=50, batch_size=64)
```


**Plot the results**

Now we visualize the results through plot graphs

```python

# Get training history metrics
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(train_acc) + 1)

# Plotting accuracy
plt.figure(figsize=(12, 6))
plt.plot(epochs, train_acc, 'b', label='Training accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Plotting loss
plt.figure(figsize=(12, 6))
plt.plot(epochs, train_loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Print final metrics
print(f"Training Accuracy: {train_acc[-1]:.4f}")
print(f"Validation Accuracy: {val_acc[-1]:.4f}")
print(f"Training Loss: {train_loss[-1]:.4f}")
print(f"Validation Loss: {val_loss[-1]:.4f}")
```
